# Statistical Test Selection Guide

Before selecting a statistical test, ask:

1. What type of data do I have?
2. What is my objective?
3. How many groups am I comparing?

# One-Sample T-Test

## Purpose

Compare a sample mean against a known benchmark.

## Formula

t = (x̄ - μ) / (s / √n)

## Variables

x̄ = Sample Mean

μ = Population Mean

s = Sample Standard Deviation

n = Sample Size

## Example

Question:

Is the average IMDb movie rating different from 6?

## When to Use

✓ Numerical Data

✓ One Sample

✓ Population SD Unknown

## Python Implementation

**INDEPENDENT T-TEST**

from scipy.stats import ttest_ind
=
drama = movie[
    movie["Genre"].str.contains("Drama", na=False)
]["Rating"]


comedy = movie[
    movie["Genre"].str.contains("Comedy", na=False)
]["Rating"]



t_stat, p_value = ttest_ind(

    drama,
    comedy,
    nan_policy="omit"
)


print("T-Statistic:", t_stat)

print("P-Value:", p_value)


if p_value < 0.05:

    print("Reject H0")

else:

    print("Fail to Reject H0")

PAIRED T-TEST


from scipy.stats import ttest_rel
=

before = [45, 50, 48, 52, 46, 49]

    print("Reject H0")

    print("Fail to Reject H0")

**F-TEST(ANOVA)**

from scipy.stats import f_oneway
=

drama = movie[
    movie["Genre"].str.contains("Drama", na=False)
]["Rating"]


comedy = movie[
    movie["Genre"].str.contains("Comedy", na=False)
]["Rating"]


action = movie[
    movie["Genre"].str.contains("Action", na=False)
]["Rating"]


f_stat, p_value = f_oneway(

    drama,

    comedy,

    action
)

print("F-Statistic:", f_stat)

print("P-Value:", p_value)

if p_value < 0.05:

    print("Reject H0")
else:

    print("Fail to Reject H0")

**CHI-SQUARE TEST**

import pandas as pd

from scipy.stats import chi2_contingency


data = {

    "Gender": ["Male","Male","Female","Female",
               "Male","Female","Male","Female"],

    "Purchase": ["Yes","No","Yes","Yes",
                 "No","No","Yes","Yes"]
}

df = pd.DataFrame(data)


table = pd.crosstab(

    df["Gender"],

    df["Purchase"]
)

chi2, p_value, dof, expected = \
chi2_contingency(table)

print("Chi-Square:", chi2)

print("P-Value:", p_value)


if p_value < 0.05:

    print("Variables are associated")
else:

    print("No significant association")

**SUMMARY TABLE**

| Test               | Purpose                   |
| ------------------ | ------------------------- |
| One-Sample T-Test  | Sample Mean vs Benchmark  |
| Independent T-Test | Compare 2 Groups          |
| Paired T-Test      | Before vs After           |
| Z-Test             | Mean Comparison (σ Known) |
| ANOVA              | Compare 3+ Groups         |
| Chi-Square         | Categorical Relationship  |
| Correlation        | Relationship Strength     |
